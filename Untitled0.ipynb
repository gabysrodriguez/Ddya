## descripción del problema

En este laboratorio se debe crear un sistema simple para organizar la atención de pacientes en un centro médico. Los pacientes pueden ser **normales** o **críticos**.

Los pacientes normales se atienden en el orden en que llegan. Sin embargo, los pacientes críticos tienen **prioridad**, por lo que siempre se atienden antes que los normales. Entre pacientes del mismo tipo, también se respeta el orden de llegada.

El sistema debe permitir **agregar pacientes** y **atender al siguiente paciente disponible**. Si no hay pacientes en espera, el sistema debe indicar que **no hay pacientes**.

## 8) Código funcional

```python
def sistema_atencion_medica(cola_normales, cola_criticos, accion):
    if accion[0] == "N":
        nombre = accion[1]
        cola_normales.append(nombre)
        print("Paciente N nuevo :", nombre)
    elif accion[0] == "C":
        nombre = accion[1]
        cola_criticos.append(nombre)
        print("Paciente C nuevo :", nombre)
    elif accion[0] == "A":
        if len(cola_criticos) > 0:
            nombre = cola_criticos[0]
            cola_criticos.pop(0)

            print("Se atiende a", nombre)
        elif len(cola_normales) > 0:
            nombre = cola_normales[0]
            cola_normales.pop(0)
            print("Se atiende a", nombre)
        else:
            print("no hay pacientes")
def main():
    cola_normales = []
    cola_criticos = []
    n = int(input('Digite las acciones'))
    if n == 0:
        return print(0)
    while n<0:
        print('el numero de acciones no puede ser negativo')
        n = int(input('Digite las acciones'))
    for i in range(n):
        accion = input('Digite la entrada:').split()
        sistema_atencion_medica(cola_normales, cola_criticos, accion)
main()
```
