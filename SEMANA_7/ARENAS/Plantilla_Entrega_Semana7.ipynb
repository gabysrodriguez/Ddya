## descripción del problema

En este laboratorio se debe crear un sistema simple para organizar la atención de pacientes en un centro médico. Los pacientes pueden ser **normales** o **críticos**.

Los pacientes normales se atienden en el orden en que llegan. Sin embargo, los pacientes críticos tienen **prioridad**, por lo que siempre se atienden antes que los normales. Entre pacientes del mismo tipo, también se respeta el orden de llegada.

El sistema debe permitir **agregar pacientes** y **atender al siguiente paciente disponible**. Si no hay pacientes en espera, el sistema debe indicar que **no hay pacientes**.
