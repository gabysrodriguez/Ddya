# **Descripción del problema**
Una empresa de distribución necesita representar la red de conexión entre sus bodegas para analizar cómo puede recorrerla desde una sede principal. Cada bodega se modela como un vértice y cada conexión entre bodegas como una arista dentro de un grafo no dirigido.

En este laboratorio, el grafo representa una red de rutas entre lugares o bodegas. El objetivo es construir la red, visualizarla mediante lista de adyacencia y matriz de adyacencia, y aplicar dos recorridos fundamentales:

BFS (Breadth First Search), que recorre por niveles usando una cola.
DFS (Depth First Search), que recorre profundizando una rama antes de retroceder, usando recursión o pila.

El propósito es comparar ambos algoritmos sobre la misma red, identificar sus diferencias y comprender en qué situaciones reales conviene usar cada uno

# **Requerimientos**

**Requerimientos funcionales**

* El sistema debe permitir registrar lugares o bodegas.
* El sistema debe permitir conectar dos lugares mediante rutas bidireccionales.
* El sistema debe mostrar la lista de adyacencia del grafo.
* El sistema debe mostrar la matriz de adyacencia del grafo.
* El sistema debe ejecutar el recorrido BFS desde un nodo inicial ingresado por el usuario.
* El sistema debe ejecutar el recorrido DFS desde un nodo inicial ingresado por el usuario.
* El sistema debe permitir agregar un nuevo lugar y conectarlo con otros dos nodos existentes.
* El sistema debe actualizar la red después de agregar un nuevo lugar.
* El sistema debe mostrar el orden de recorrido obtenido tanto en BFS como en DFS.

**Requerimientos no funcionales**

* El programa debe estar implementado únicamente en Python.
* El código debe ser legible, modular y fácil de mantener.
* La solución debe evitar recorridos repetidos sobre nodos ya visitados.
* La ejecución debe ser eficiente para grafos pequeños y medianos.
* La interacción debe realizarse por consola con mensajes claros.
* Los vecinos deben explorarse en orden alfabético para que el recorrido sea reproducible.

# **Historias de usuario**

* **Historia de usuario 1**

  Como administrador de la red de bodegas,quiero registrar lugares y rutas, para representar correctamente la estructura de conexiones de la empresa.

  **Criterios de aceptación**

  Dado un conjunto de lugares, el sistema los almacena en el grafo.Dadas dos bodegas válidas, el sistema crea una conexión bidireccional.
  Al finalizar, el sistema muestra la lista de adyacencia actualizada.
   
* **Historia de usuario 2**
  Como operador logístico, quiero ejecutar BFS desde una bodega principal, para conocer el orden de exploración por niveles.

  **Criterios de aceptación**

  Dado un nodo inicial existente, el sistema ejecuta BFS.
  El recorrido visita primero los vecinos directos.
  El sistema muestra el orden final del recorrido.

* **Historia de usuario 3**
  Como analista de red,quiero ejecutar DFS desde una bodega principal, para explorar profundamente una ruta antes de retroceder.

  **Criterios de aceptación**

  Dado un nodo inicial válido, el sistema ejecuta DFS.
  El sistema marca los nodos visitados para evitar ciclos infinitos. El recorrido se muestra en pantalla en el orden de visita.

  **Historia de usuario 4**

  Como usuario del sistema, quiero agregar una nueva bodega a la red, para actualizar el grafo cuando la empresa crezca.

  **Criterios de aceptación**

  El sistema permite ingresar el nombre del nuevo lugar. El sistema solicita dos conexiones para integrarlo a la red.
  Luego de agregarlo, el sistema actualiza la lista y la matriz de adyacencia.


**Diagrama de flujo**

El proceso inicia en el menú principal. Desde allí el usuario puede crear la red, ejecutar BFS, ejecutar DFS, agregar un lugar o salir. Cada opción dirige a un proceso específico. BFS utiliza una cola para recorrer el grafo por niveles, mientras que DFS profundiza una rama mediante recursión.
